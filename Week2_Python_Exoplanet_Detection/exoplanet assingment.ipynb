# GJ 8999 b Exoplanet Characterization

## Introduction
In this assignment, we will estimate the mass and radius of the exoplanet GJ 8999 b, which is orbiting the M dwarf star GJ 8999.

### Star Properties
- **Mass of GJ 8999**: \( M_{\text{star}} = 0.2 \, M_{\odot} \)
- **Radius of GJ 8999**: \( R_{\text{star}} = 0.2 \, R_{\odot} \)

## 1. Inclination of GJ 8999 b
The inclination \( i \) can be calculated using the formula:

\[
i = \arccos\left(\frac{R_{\text{star}}}{R_{\text{star}} + R_{\text{planet}}}\right)
\]

### 1.1 Transit Depth
The transit depth can be expressed as:

\[
\text{Transit Depth} = \left(\frac{R_{\text{planet}}}{R_{\text{star}}}\right)^2
\]

From the transit depth, we can estimate the radius of the planet \( R_{\text{planet}} \).

## 2. Radial Velocity Data
The mass of the planet can be derived from radial velocity data using the equation:

\[
M_{\text{planet}} = \left(\frac{K}{\sin(i)}\right)^3 \frac{P^2}{(2\pi G)^{1/2}}
\]

Where:
- \( K \) is the radial velocity amplitude
- \( P \) is the orbital period
- \( G \) is the universal gravitational constant

## Conclusion
Using the above formulas and data, we calculated the mass and radius of GJ 8999 b and analyze the results.
